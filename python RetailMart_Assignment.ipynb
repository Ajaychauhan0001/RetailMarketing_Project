```python
import pandas as pd
import numpy as np
```

```python
# Load CSV files
sales_df = pd.read_csv("data/sales_data.csv")
products_df = pd.read_csv("data/products.csv")
stores_df = pd.read_csv("data/stores.csv")
```

```python
print("===== SHAPE OF DATAFRAMES =====")
print("Sales Data :", sales_df.shape)
print("Products Data :", products_df.shape)
print("Stores Data :", stores_df.shape)
```

```python
print("\n===== SALES DATA =====")
print(sales_df.head())

print("\n===== PRODUCTS DATA =====")
print(products_df.head())

print("\n===== STORES DATA =====")
print(stores_df.head())
```

```python
print("\n===== MISSING VALUES =====")

print("\nSales Data")
print(sales_df.isnull().sum())

print("\nProducts Data")
print(products_df.isnull().sum())

print("\nStores Data")
print(stores_df.isnull().sum())
```

```python
print("\n===== DATA CLEANING =====")

# Count duplicates
duplicate_count = sales_df.duplicated().sum()

print(f"Duplicate Rows Found: {duplicate_count}")

# Remove duplicates
sales_df = sales_df.drop_duplicates()

print(f"Shape after removing duplicates: {sales_df.shape}")
```

```python
# Fill missing quantity with 0
sales_df["quantity"] = sales_df["quantity"].fillna(0)
```

```python
# Drop rows where amount is NULL
sales_df = sales_df.dropna(subset=["amount"])

print("Shape after removing NULL amount rows:", sales_df.shape)
```

```python
# Convert sale_date to datetime
sales_df["sale_date"] = pd.to_datetime(sales_df["sale_date"])

# Convert amount to float
sales_df["amount"] = sales_df["amount"].astype(float)
```

```python
print("\n===== DATA TYPES =====")
print(sales_df.dtypes)
```

```python
print("\n===== CLEANED SALES DATA =====")
print(sales_df.head())
```

```python
print("\n===== DATA TRANSFORMATION =====")

# Merge sales and products
merged_df = pd.merge(sales_df, products_df, on="product_id", how="inner")

# Merge with stores
merged_df = pd.merge(merged_df, stores_df, on="store_id", how="inner")

print("\n===== MERGED DATA =====")
print(merged_df.head())
```

```python
merged_df["total_revenue"] = merged_df["quantity"] * merged_df["price"]
```

```python
print("\n===== REVENUE STATISTICS =====")

print("Mean Revenue :", np.mean(merged_df["total_revenue"]))
print("Maximum Revenue :", np.max(merged_df["total_revenue"]))
print("Minimum Revenue :", np.min(merged_df["total_revenue"]))
```

```python
city_revenue = (
    merged_df.groupby("city")["total_revenue"]
    .sum()
    .sort_values(ascending=False)
)

print("\n===== TOTAL REVENUE PER CITY =====")
print(city_revenue)
```

```python
print("\n===== FINAL MERGED DATA =====")
print(merged_df)
```

```python
import sqlite3
```

```python
print("\n===== LOADING DATA INTO SQLITE DATABASE =====")

# Create SQLite database
conn = sqlite3.connect("database/retail_sales.db")

# Load DataFrame into SQLite table
merged_df.to_sql(
    "retail_sales",
    conn,
    if_exists="replace",
    index=False
)

print("Data loaded successfully into SQLite database.")
```

```python
query1 = """
SELECT
    product_name,
    SUM(quantity) AS total_quantity
FROM retail_sales
GROUP BY product_name
ORDER BY total_quantity DESC
LIMIT 3;
"""

top_products = pd.read_sql(query1, conn)

print("\n===== TOP 3 BEST-SELLING PRODUCTS =====")
print(top_products)
```

```python
query2 = """
SELECT
    store_name,
    sale_date,
    SUM(total_revenue) AS total_revenue
FROM retail_sales
GROUP BY store_name, sale_date
ORDER BY sale_date;
"""

store_revenue = pd.read_sql(query2, conn)

print("\n===== TOTAL REVENUE PER STORE PER DAY =====")
print(store_revenue)
```

```python
conn.close()
```

```python
print("\n===== SUMMARY REPORT =====")

# Total transactions
total_transactions = len(merged_df)

# Total revenue
total_revenue = merged_df["total_revenue"].sum()

# Top selling city
top_city = (
    merged_df.groupby("city")["total_revenue"]
    .sum()
    .idxmax()
)

# Top selling product
top_product = (
    merged_df.groupby("product_name")["quantity"]
    .sum()
    .idxmax()
)

print(f"Total Transactions : {total_transactions}")
print(f"Total Revenue      : {total_revenue}")
print(f"Top Selling City   : {top_city}")
print(f"Top Selling Product: {top_product}")
```
